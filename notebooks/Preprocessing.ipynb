```python
# Importing necessary libraries
import numpy as np
import pandas as pd
```

### Read file


```python
#reading the .csv file and declare datatype to avoid low memory issues
df = pd.read_csv("trec07.csv",dtype='unicode')
```

### Handling Null Values

```python
# replacing na values
df["Return-Path"].fillna("unknown", inplace = True) 
df["Message-ID"].fillna("unknown", inplace = True) 
df["From"].fillna("unknown", inplace = True) 
df["Reply-To"].fillna("unknown", inplace = True) 
df["To"].fillna("unknown", inplace = True) 
df["Subject"].fillna("unknown", inplace = True) 
df["Date"].fillna("unknown", inplace = True) 
df["X-Mailer"].fillna("unknown", inplace = True)
df["MIME-Version"].fillna("0", inplace = True)
df["Content-Type"].fillna("unknown", inplace = True)
df["X-Priority"].fillna("0", inplace = True)
df["X-MSMail-Priority"].fillna("0", inplace = True)
df["Status"].fillna("0", inplace = True)
df["Content-Length"].fillna("0", inplace = True)
df["Content-Transfer-Encoding"].fillna("unknown", inplace = True)
df["Lines"].fillna("0", inplace = True)
df
```

```python
#check if there is any other null values
df.isnull().sum()
```

```python
#removing duplicates and removing \n characters and re
df.drop_duplicates()
df = df.replace("\n","",regex = True)
df = df.replace("re :","",regex = True)
```
